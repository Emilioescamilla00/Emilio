```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
sns.boxplot(data = df['peso']
           )
```

```python
df['peso'].plot(kind='line')
```

```python
df['peso'].plot(kind='hist')
```

```python
sns.boxplot(data = df['cintura']
           )
```

```python
df['cintura'].plot(kind='line')
```

```python
df['cintura'].plot(kind='hist')
```

```python
df.columns
```

```python
sns.boxplot(data = df.loc[df['sexo'] == '1', 
                          ['peso', 'cintura']],
           ).set_title('Box plot de Sexo 1')
```

```python
sns.lmplot (x = 'peso', 
             y = 'cintura',
             data = df
            )

plt.show()
```

```python
sns.lmplot (x = 'peso', 
             y = 'cintura',
             data = df,
            fit_reg = False,
            hue = 'sexo'
            )
plt.title('Scatter plot of peso vs cintura', fontsize = 17)
plt.xticks(fontsize = 15, rotation = 45)
plt.yticks(fontsize = 15)
plt.show()
```
