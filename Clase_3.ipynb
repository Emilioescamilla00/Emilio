```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME', 'CTYNAME'], #aqui puedes poner lo que quieras, si es uno se pone '__', si son más se pone ['__', '__']
                          values = 'BIRTHS2010',
                          aggfunc = ['std', 'sum', 'mean', 'median'] #aqui puedes poner lo que estes buscando, sum, mean, median, std... en tipo '___' o si quieres varias ['_', '_']
                         )
my_pivot
```
