```python
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = 'STNAME', #aqui puedes poner lo que quieras, si es uno se pone '__', si son más se pone ['__', '__']
                          values = 'BIRTHS2010',
                          aggfunc = ['std', 'sum', 'mean', 'median'] #aqui puedes poner lo que estes buscando, sum, mean, median, std... en tipo '___' o si quieres varias ['_', '_']
                         )
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME', 'CTYNAME'], #aqui puedes poner lo que quieras, si es uno se pone '__', si son más se pone ['__', '__']
                          values = ['BIRTHS2010', 'BIRTHS2011'], # tambien se puede aqui hacer lo mismo para tener más variables
                          #aggfunc = ['std', 'sum', 'mean', 'median'] #aqui puedes poner lo que estes buscando, sum, mean, median, std... en tipo '___' o si quieres varias ['_', '_']
                          aggfunc = 'median',
                          columns = 'SUMLEV',
                          margins = True
                         )
my_pivot
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/airline_data.csv', low_memory=False)
df.head()
```

```python
df.columns
```

```python
df.shape #nos da la cantidad de columnas y filas
```

```python
columnz = df.columns.tolist() #sirve para ponerlas en lista
columnz.sort() #sirve para ponerlas en orden alfabetico
columnz
```

```python
delay_pivot = pd.pivot_table(df,
                             index = ['Year', 'Quarter'],
                             values = 'DepDelay',
                             aggfunc = 'sum',
                             columns = 'DepartureDelayGroups',
                            )
delay_pivot
```
