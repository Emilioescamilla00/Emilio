### loc y iloc

[row_inicial:row_final, col_inicial:col_final]

si usas loc, usa nombres de los renglones y nombres de las columnas

si usas iloc, usa indices de los renglones y de las columnas

```python
df.loc[2:7, 'Age':'Height']
```

```python
df.iloc[2:10, 3:6]
```
