```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
%ls datasets/
```

```python
df = pd.read_excel('datasets/mic.xlsx',
                 engine = 'openpyxl'
                )
df.head()
```

```python
counts = df['Age'].value_counts()
#print(type(counts))
counts = counts.to_frame()
#print(type(counts))
#print(counts)
counts['%'] = (counts.loc[:, 'Age']/counts.loc[:, 'Age'].sum(axis = 0))*100
counts
```

```python
counts = counts.reindex(['<=20', '21 - 30', '31 - 40', '41 - 50'])
counts
```

```python
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index': 'Age_group'},
                       axis = 1
                      )
counts
```

```python
color_list = ['gold', 'yellowgreen', 'coral', 'skyblue']
explode_list = [0.03, 0.04, 0.06, 0.05]

counts['Age'].plot(kind = 'pie',
                   figsize = (15, 6),
                   startangle = 90,
                   labels = None,
                   autopct = '%1.1f%%,',
                   pctdistance = 1.1,
                   colors = color_list,
                   explode = explode_list,
                   shadow = True
                  )
plt.legend(labels = counts.Age_group,
          bbox_to_anchor = (1.0, 1.0),
           title = 'Legend',
           title_fontsize = 13
          )
plt.title('Pie chart Age_group', fontsize = 13)
plt.show()
```

```python
x = list(counts['Age_group'])
y = list(counts['Age'])

plt.bar(x, y, width = 0.7, color = 'darkcyan')
plt.title('Grafica de barras de Age_group', fontsize = 14)

plt.show
```

```python
plt.subplot(1,2,1)

x = list(counts['Age_group'])
y = list(counts['Age'])

plt.title('Grafica de barras de Age_group', fontsize = 14)
plt.bar(x, y, width = 0.7, color = 'darkcyan')

plt.show

plt.subplot(1,2,2)

color_list = ['gold', 'yellowgreen', 'coral', 'skyblue']
explode_list = [0.03, 0.04, 0.06, 0.05]

counts['Age'].plot(kind = 'pie',
                   figsize = (15, 6),
                   startangle = 90,
                   labels = None,
                   autopct = '%1.1f%%,',
                   pctdistance = 1.1,
                   colors = color_list,
                   explode = explode_list,
                   shadow = True
                  )
plt.legend(labels = counts.Age_group,
          bbox_to_anchor = (1.0, 1.0),
           title = 'Legend',
           title_fontsize = 13
          )
plt.title('Pie chart Age_group', fontsize = 13)

plt.show()
```
